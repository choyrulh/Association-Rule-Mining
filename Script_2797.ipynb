```python
from mlxtend.frequent_patterns import apriori, association_rules
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/media/irul/DATA/grosir.csv', header = None)
```

```python
transaksi = []
for i in range(0, 9836):
  transaksi.append([str(df.values[i,j]) for j in range(0, 32)])
```

```python
transaksi
```

```python
association_rules=apriori (transaksi, min_support=0.0035, min_confidence=0.2, min_lift=3, miin_length=2 )
```

```python
association_rules=list(association_rules)
```

```python
for item in association_rules:
    pair = item[0]
    items = [x for x in pair]
    print(items)
    print("Support: " + str(item[1]))

    print("Confidence: " + str(item[2][0][2]))
    print("Lift: " + str(item[2][0][3]))
    print("============================")
```
